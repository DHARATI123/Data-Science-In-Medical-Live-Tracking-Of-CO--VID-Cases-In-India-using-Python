```
import pandas as pd
df = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
```

```
df
```

```
df = df.iloc[:-1,1:]
```

```
df.head(-5)
```

```
import matplotlib.pyplot as plt
import numpy as np
total_states = np.arange(len(df['state_name']))
```

```
total_states
```

```
df['active']
```

```
max(df['positive'])
```

```
#@title total positive cases based on states in india 

```

```
from matplotlib.pyplot import figure
figure(num=None, figsize = (10,8) , dpi = 80, facecolor ='w', edgecolor = 'k')
plt.barh(total_states,df['positive'],align ='center',alpha = 0.5,
         color = (1,0,0), 
         edgecolor = (0.5,0.2,0.8))
plt.yticks(total_states,df['state_name'])
plt.xlim(1,max(df['positive'])+10)
plt.xlabel('positive no of cases')
plt.ylabel('corona virus cases')
plt.show()
```

```
#@title Active no. of cases based on states
```

```
from matplotlib.pyplot import figure
figure(num=None, figsize = (10,8) , dpi = 80, facecolor ='w', edgecolor = 'k')
plt.barh(total_states,df['active'],align ='center',alpha = 0.5,
         color = (1,0.5,0), 
         edgecolor = (0.5,0.2,0.8))
plt.yticks(total_states,df['state_name'])
plt.xlim(1,max(df['active'])+10)
plt.xlabel('active no of cases')
plt.ylabel('corona virus cases')
plt.show()
```

```
#@title Death no of cases
```

```
from matplotlib.pyplot import figure
figure(num=None, figsize = (10,8) , dpi = 80, facecolor ='w', edgecolor = 'k')
plt.barh(total_states,df['death'],align ='center',alpha = 0.5,
         color = (0,0,1), 
         edgecolor = (0.5,0.2,0.8))
plt.yticks(total_states,df['state_name'])
plt.xlim(1,max(df['death'])+10)
plt.xlabel('Death no of cases')
plt.ylabel('corona virus cases')
plt.show()
```
